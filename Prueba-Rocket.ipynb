a) ¿Cómo obtendrías la información más reciente de los usuarios?

Primero importaríamos la librería de pandas para poder leer el Excel que previamente se nos fue entregado (puede ser de tipo json, csv, etc), posteriormente tendríamos que verificar que la columna de fechas que queremos manejar sea de tipo Timestamp, si no es el caso lo podemos convertir, pero en este caso si es de este tipo, así que podemos usar el método sort_values para ordenar las fechas, pero le tenemos que indicar en los parámetros el nombre de la columna y en el otro como serán ordenados, por defecto siempre se nos ordena de la fecha mas antigua a la mas reciente, es por eso que colocamos ascending=False para que se nos ordene las fechas de la mas reciente a las mas antigua.

```python
import pandas as pd

# importamos nuestro excel con el metodo de pandas read_excel
#df_usuarios = pd.read_excel(r"C:\Users\misae\Desktop\Rocket\Ejemplo usuarios.xlsx") esta se utiliza en un entorno local

#converti el archivo excel a tipo csv para posteriormente subirlo a GitHub con el fin de que al momento de correr 
#el codigo en cualquier computadora corra sin necesidad de importar el excel

df_usuarios = pd.read_csv("https://raw.githubusercontent.com/Misael-Alvarez/Rocket/main/Ejemplo%20usuarios.csv")
# previamente se verifico que sean de tipo Timestamp los elementos de la columna fecha_respuesta
# ordeamos con el metodo sort de la mas reciente a la mas antigua
df_usuarios_ordenados = df_usuarios.sort_values(by='fecha_respuesta',ascending=False).copy() # creamos una copia con los datos ordenados
df_usuarios_ordenados
```
